```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'date' column to datetime format
df2['date'] = pd.to_datetime(df2['date'])

# Extract year and month from the 'date' column
df2['Year'] = df2['date'].dt.year
df2['Month'] = df2['date'].dt.month_name()

# Group by Year and Month, count the number of reviews, and reset the index
reviews_by_month = df2.groupby(['Year', 'Month']).size().reset_index(name='Number of Reviews')

# Bar plot for the number of reviews by month
plt.figure(figsize=(14, 8))
sns.barplot(x='Number of Reviews', y='Month', hue='Year', data=reviews_by_month, palette='viridis')
plt.title('Number of Reviews by Month and Year')
plt.xlabel('Number of Reviews')
plt.ylabel('Month')
plt.show()
```

```python
print(df2.head())
```

```python
# Number of Positive Reviews by Month for Each Year

df2['Year'] = df2['date'].dt.year
df2['Month'] = df2['date'].dt.month_name()

df_positive = df2[df2['Sentiment'] == 'positive']

plt.figure(figsize=(14, 8))
sns.countplot(x='Month', hue='Year', data=df_positive.sort_values(['Year', 'date']))

for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Number of Positive Reviews by Month for Each Year')
plt.xlabel('Month')
plt.ylabel('Number of Positive Reviews')
plt.legend(title='Year')
plt.show()
```

```python
# Number of Positive Reviews by Month in 2019

import matplotlib.pyplot as plt

df2['Year'] = df2['date'].dt.year
df2['Month'] = df2['date'].dt.month_name()

df_2019_positive = df2[(df2['Year'] == 2019) & (df2['Sentiment'] == 'positive')]

plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=df_2019_positive.sort_values('date'))
plt.title('Number of Positive Reviews by Month in 2019')
plt.xlabel('Month')
plt.ylabel('Number of Positive Reviews')
plt.show()
```

```python
# Number of Reviews by Location

df2['Year'] = df2['date'].dt.year
reviews_by_location = df2.groupby('Location').size().reset_index(name='Number of Reviews')
reviews_by_location = reviews_by_location.sort_values(by='Number of Reviews', ascending=False)

plt.figure(figsize=(14, 8))
sns.barplot(x='Number of Reviews', y='Location', data=reviews_by_location.head(15))
plt.title('Number of Reviews by Location')
plt.xlabel('Number of Reviews')
plt.ylabel('Location')
plt.show()
```

```python
# Sentiment Distribution Year-Wise

df2['Year'] = df2['date'].dt.year

sentiment_percentage = df2.groupby(['Year', 'Sentiment']).size() / df2.groupby('Year').size() * 100
sentiment_percentage = sentiment_percentage.unstack()

plt.figure(figsize=(12, 6))
sentiment_percentage.plot(kind='bar', stacked=True, colormap='viridis', edgecolor='black')

for i, year in enumerate(sentiment_percentage.index):
    total_height = 0
    for sentiment, value in sentiment_percentage.loc[year].items():
        plt.text(i, total_height + value / 2, f'{value:.2f}%', ha='center', va='center', color='white', fontweight='bold')
        total_height += value

plt.title('Sentiment Distribution Year-Wise')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend(title='Sentiment')
plt.show()
```

```python
#Overall Sentiment Distribution

overall_sentiment_percentage = df2['Sentiment'].value_counts(normalize=True) * 100

print("Overall Sentiment Distribution:")
print(overall_sentiment_percentage)
```

```python
# Sentiment distribution year-wise in percentage
sentiment_percentage_yearwise = df2.groupby(['Year', 'Sentiment']).size() / df2.groupby('Year').size() * 100

#year-wise in numbers
sentiment_counts_yearwise = df2.groupby(['Year', 'Sentiment']).size()

#month-wise of each year
sentiment_monthwise = df2.groupby(['Year', 'Month', 'Sentiment']).size() / df2.groupby(['Year', 'Month']).size() * 100

#location-wise
sentiment_locationwise = df2.groupby(['Location', 'Sentiment']).size() / df2.groupby('Location').size() * 100

print("Sentiment Distribution Year-Wise in Percentage:")
print(sentiment_percentage_yearwise)

print("\nSentiment Distribution Year-Wise in Numbers:")
print(sentiment_counts_yearwise)

print("\nSentiment Distribution Month-Wise of Each Year:")
print(sentiment_monthwise)

print("\nSentiment Distribution Location-Wise:")
print(sentiment_locationwise)
```

```python
# Average Review Length by Sentiment

import pandas as pd

df2['Review Length'] = df2['Review'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)

# Group by 'Sentiment' and calculate the average review length
average_length_by_sentiment = df2.groupby('Sentiment')['Review Length'].mean()

# Display the results
print("Average Review Length by Sentiment:")
print(average_length_by_sentiment)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate percentage of reviews by sentiment
sentiment_percentage = df2['Sentiment'].value_counts(normalize=True) * 100

# Bar plot for sentiment distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_percentage.index, y=sentiment_percentage.values, palette='pastel')

# Annotate the bars with percentage values
for index, value in enumerate(sentiment_percentage):
    plt.text(index, value + 1, f'{value:.2f}%', ha='center', va='bottom')

plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Percentage of Reviews')
plt.show()
```

```python
# Filter only positive reviews
positive_reviews = df2[df2['Sentiment'] == 'positive']

# Group by location and count the number of positive reviews
positive_reviews_by_location = positive_reviews.groupby('Location').size().reset_index(name='Number of Positive Reviews')

# Sort the DataFrame by the number of positive reviews in descending order
positive_reviews_by_location = positive_reviews_by_location.sort_values(by='Number of Positive Reviews', ascending=False)

# Bar plot of the number of positive reviews by location
plt.figure(figsize=(14, 8))
sns.barplot(x='Number of Positive Reviews', y='Location', data=positive_reviews_by_location.head(15), palette='viridis')

# Annotate the bars with the number of positive reviews
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_width())}', (p.get_width(), p.get_y() + p.get_height() / 2.), ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.title('Most Loved Locations Based on Positive Reviews')
plt.xlabel('Number of Positive Reviews')
plt.ylabel('Location')
plt.show()
```

   - **Conclusion:** Negative reviews tend to have a longer average length compared to neutral and positive reviews, suggesting that customers expressing dissatisfaction may provide more detailed feedback.

6. **Most Loved Location - USA:**
